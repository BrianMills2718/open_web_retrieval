# Retrieval Control Surface and Behavior Verification

Status: planned

This notebook is the planning artifact for expanding the shared `SearchQuery` contract so consumers can declare retrieval depth, detail, guidance, and corpus intent, and for proving the adapters honor those controls.


## Inputs

- `src/open_web_retrieval/models.py`
- `src/open_web_retrieval/adapters/tavily.py`
- `src/open_web_retrieval/adapters/exa.py`
- `tests/test_adapters.py`
- `tests/test_client.py`


## Contract Decisions

1. Add typed controls to `SearchQuery` rather than consumer-local wrappers.
2. Keep the contract generic: depth, detail, guidance, and corpus intent.
3. Verify behavior with transport-capture adapter tests that inspect the provider request bodies.


## Acceptance Criteria

- Tavily no longer hardcodes search depth when the query declares one.
- Exa no longer hardcodes one deep/highlights shape when the query declares otherwise.
- The shared contract remains typed and consumer-expressive without a free-form options bag.
